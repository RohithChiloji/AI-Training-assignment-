```python
def grade(scores):
    if     90< scores <= 100:
        return 'S'
    elif   80 < scores <= 90:
        return 'A+'
    elif   70< scores <= 80:
        return 'A'
    elif   60 < scores <= 70:
        return 'A+'
    elif   50 < scores <= 60:
        return 'B+'
    else:
        return 'F'
```
